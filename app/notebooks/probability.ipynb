```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# 
n = 10000  # 

# ======== 1. Uniform Distribution========
# : 
# : 
uniform_samples = np.random.uniform(0, 1, n)  # 01

# ======== 2. Normal/Gaussian Distribution========
# : 
# : 
# : ()()
mu, sigma = 0, 1  # 01
normal_samples = np.random.normal(mu, sigma, n)

# ======== 3. Exponential Distribution========
# : 
# : 
# : ()1/
lambda_param = 1.5
exp_samples = np.random.exponential(1/lambda_param, n)

# ======== 4. Binomial Distribution========
# : n
# : 
p = 0.3  # 
trials = 20  # 
binom_samples = np.random.binomial(trials, p, n)

# ========  ========
fig, axes = plt.subplots(4, 1, figsize=(12, 16))

# 
axes[0].hist(uniform_samples, bins=30, alpha=0.7, density=True, label='')
x_uniform = np.linspace(0, 1, 100)
axes[0].plot(x_uniform, stats.uniform.pdf(x_uniform, 0, 1), 'r-', lw=2, label='PDF')
axes[0].set_title(' U(0,1)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# 
axes[1].hist(normal_samples, bins=30, alpha=0.7, density=True, label='')
x_normal = np.linspace(-4, 4, 100)
axes[1].plot(x_normal, stats.norm.pdf(x_normal, mu, sigma), 'r-', lw=2, label='PDF')
axes[1].set_title(f' N({mu},{sigma})')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# 
axes[2].hist(exp_samples, bins=30, alpha=0.7, density=True, label='')
x_exp = np.linspace(0, 5, 100)
axes[2].plot(x_exp, stats.expon.pdf(x_exp, scale=1/lambda_param), 'r-', lw=2, label='PDF')
axes[2].set_title(f' (={lambda_param})')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

# 
x_binom = np.arange(0, trials+1)
axes[3].hist(binom_samples, bins=np.arange(0, trials+2)-0.5, alpha=0.7, density=True, label='')
axes[3].plot(x_binom, stats.binom.pmf(x_binom, trials, p), 'ro-', lw=2, label='PMF')
axes[3].set_title(f' B({trials},{p})')
axes[3].legend()
axes[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ========  ========
# 
print("=====  U(0,1) =====")
print(f": {0.5}")
print(f": {np.mean(uniform_samples):.4f}")
print(f": {1/12}")
print(f": {np.var(uniform_samples):.4f}")
print()

print("=====  N(0,1) =====")
print(f": {mu}")
print(f": {np.mean(normal_samples):.4f}")
print(f": {sigma**2}")
print(f": {np.var(normal_samples):.4f}")
print()

print("=====  (={lambda_param}) =====")
print(f": {1/lambda_param}")
print(f": {np.mean(exp_samples):.4f}")
print(f": {1/lambda_param**2}")
print(f": {np.var(exp_samples):.4f}")
print()

print("=====  B({trials},{p}) =====")
print(f": {trials*p}")
print(f": {np.mean(binom_samples):.4f}")
print(f": {trials*p*(1-p)}")
print(f": {np.var(binom_samples):.4f}")
```
